```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x + 3)**2

def df(x):
    return 2 * (x + 3)

def gradient_descent(start_x, learning_rate, num_iterations):
    x = start_x
    x_history = [x]
    
    print(f"Starting Gradient Descent at x = {x}\n")
    for i in range(num_iterations):

        gradient = df(x)
        x = x - (learning_rate * gradient)
        x_history.append(x)

        if (i + 1) % 10 == 0:
            print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")
            
    print(f"\nGradient Descent finished after {num_iterations} iterations.")
    return x, x_history

start_point = 2       # As requested
learn_rate = 0.1      # A good learning rate
iterations = 50       # Number of steps to take

min_x, history = gradient_descent(start_point, learn_rate, iterations)
```

```python
x_vals = np.linspace(-10, 4, 100)
y_vals = f(x_vals)
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label = 'Graph')
y_history = [f(x) for x in history]
plt.plot(history, y_history, 'ro-', label="Gradient Descent Path")
plt.scatter(history[-1] , y_history[-1] , s = 100, color = 'green')
plt.title('Gradient Descent Visualization')
plt.xlabel("x")
plt.ylabel("y = f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
