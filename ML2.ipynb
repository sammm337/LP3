```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv('emails.csv')
```

```python
df.head()
```

```python
df = df.drop(['Email No.'], axis = 1)
X = df.drop(columns = ['Prediction'])
y = df['Prediction']
```

**Train Test Split**

```python
X_train , X_test, y_train, y_test = train_test_split(X,y , test_size = 0.3, random_state = 42)
```

**Scaling**

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**KNN Classification**

```python
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train_scaled,y_train)
knn_prediction = knn.predict(X_test_scaled)
print(f'Accuracy of KNN model is : {accuracy_score(y_test , knn_prediction) * 100 : .2f} %' )
print(classification_report(y_test, knn_prediction))
```

**Support Vector Machine**

```python
svm = SVC(random_state = 42, kernel = 'linear')
svm.fit(X_train_scaled , y_train)
svm_prediction = svm.predict(X_test_scaled)
print(f'Accuracy of SVM model is : {accuracy_score(y_test , svm_prediction) * 100 : .2f} %' )
print(classification_report(y_test, svm_prediction))
```
